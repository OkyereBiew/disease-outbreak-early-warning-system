## Loading the Dataset

The dengue forecasting dataset consists of environmental measurements and weekly dengue case counts.

The environmental variables include rainfall, humidity, temperature, and vegetation indices that may influence disease transmission dynamics.


```python
features = pd.read_csv(
    "/content/dengue_features_train.csv"
)

labels = pd.read_csv(
    "/content/dengue_labels_train.csv"
)

print("Features Shape:", features.shape)
print("Labels Shape:", labels.shape)
```

## Inspecting the Data Structure

Before merging the datasets, it is important to understand the available variables and verify that the datasets contain the expected observations.


```python
features.head()
```

```python
labels.head()
```

## Dataset Integration

The environmental features and dengue case counts are merged using the common identifiers:

* City
* Year
* Epidemiological Week

This creates a single dataset for analysis.


```python
dengue_df = pd.merge(
    features,
    labels,
    on=[
        "city",
        "year",
        "weekofyear"
    ]
)

dengue_df.head()
```

```python
print(
    "Merged Dataset Shape:",
    dengue_df.shape
)
```

## Data Quality Assessment

Missing values are common in epidemiological and environmental surveillance systems.

Understanding the extent of missingness helps guide preprocessing decisions in later stages of the project.


```python
missing_values = (
    dengue_df.isnull()
    .sum()
    .sort_values(
        ascending=False
    )
)

missing_values[
    missing_values > 0
]
```
